```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.describe()
```

```python
x_train = train[['pressure', 'maxtemp', 'temparature', 'mintemp', 'dewpoint', 'humidity', 'cloud', 'sunshine', 'winddirection', 'windspeed']]
y_train = train['rainfall']
```

```python
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
x_test = test[['pressure', 'maxtemp', 'temparature', 'mintemp', 'dewpoint', 'humidity', 'cloud', 'sunshine', 'winddirection', 'windspeed']]
```

```python
dt_probs = model.predict_proba(x_test)[:, 1] 
```

```python
len(dt_probs)
```

```python
res = pd.DataFrame({
    'id': test['id'],
    'rainfall': dt_probs
})
```

```python
res.to_csv('predictions-dt.csv', index=False)
```
